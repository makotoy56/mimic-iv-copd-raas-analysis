## 4. Statistical Methods
SAS and Python parameter outputs are merged by `Variable`, and absolute and relative differences are calculated.

---


## 5. Results
### 5.1 Load Parameter Outputs


```python
import pandas as pd
import numpy as np

sas_params = pd.read_csv("../sas/outputs/sas_secondary_logistic_validation_parameters.csv")
python_params = pd.read_csv("../python/outputs/python_logistic_parameters.csv")
```

### 5.2 Compare Parameter Estimates


```python
comparison = sas_params[["Variable", "Estimate"]].merge(
    python_params[["Variable", "Python_Estimate"]],
    on="Variable",
    how="inner",
)

comparison = comparison.rename(columns={"Estimate": "SAS_Estimate"})
comparison["Absolute difference"] = (
    comparison["Python_Estimate"] - comparison["SAS_Estimate"]
).abs()
comparison["Relative difference (%)"] = np.where(
    comparison["SAS_Estimate"] != 0,
    comparison["Absolute difference"] / comparison["SAS_Estimate"].abs() * 100,
    np.nan,
)

comparison.sort_values("Variable").reset_index(drop=True)
```

---

## 6. Discussion
Agreement between exported parameter estimates supports reproducibility of the model implementation across SAS and Python. This notebook only compares precomputed parameter outputs.
